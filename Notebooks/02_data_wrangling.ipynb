## Derive State-wide Summary Statistics For Our Market Segment

```python
#Code task 27#
#Add named aggregations for the sum of 'daysOpenLastYear', 'TerrainParks', and 'NightSkiing_ac'
#call them 'state_total_days_open', 'state_total_terrain_parks', and 'state_total_nightskiing_ac',
#respectively
#Finally, add a call to the reset_index() method (we recommend you experiment with and without this to see
#what it does)
state_summary = ski_data.groupby('state').agg(
    resorts_per_state=pd.NamedAgg(column='Name', aggfunc='size'), #could pick any column here
    state_total_skiable_area_ac=pd.NamedAgg(column='SkiableTerrain_ac', aggfunc='sum'),
    state_total_days_open=pd.NamedAgg(column='daysOpenLastYear', aggfunc='sum'),
    state_total_terrain_parks=pd.NamedAgg(column='TerrainParks', aggfunc='sum'),
    state_total_nightskiing_ac=pd.NamedAgg(column='NightSkiing_ac', aggfunc='sum')
).reset_index()
state_summary.head()
```

## Summary

We're given a dataset that describes 330 resorts in the US. Our higher level goal is to make a pricing model that can predict what the price would be for a resort in Montana. The goal of this notebook is to familiarize ourselves with the data and organize it. An initial look tells us there's three categorical columns (Name,Region,State), two possible target feature columns (Adult Weekend Price, Adult Weekday Price) and the rest of the 27 columns are numeric. Our own resort is also included in the dataset which is odd since we'd be using our own resort and price to predict/decide what our price should be. Also found that many of the numeric columns including the target features had missing values.

Tackling the categorical columns first, I used them to confirm that every row represents a unique resort. Both Region and State columns are very similar.From both it's more likely that I'd use state for modelling and so set up a seperate state-wide summary statistics table to help encode state-level data before dropping rows. Most of the state level data was aggregated from the dataset itself such as resorts per state and total skiable area but two of those columns, state population and state area was scraped from Wikipedia. 

Next I looked at the distributions for the numeric columns. The 'fastEight' column was dropped immediately as half the values were missing and the other half were the same. Also removed or fixed some outliers in Skiable Terrain, Snow making and YearsOpen.
Removed rows with no price data at all and then reviewed distributions for all the numeric columns. 'Trams','FastSixes' and 'FastQuads' had very low variance and will have to keep an eye out for them.

Now I looked at the relationship between 'AdultWeekend' and 'AdultWeekday'. Since they had a strong linear relationship, it was decided that the feature with the least number of missing values would be the target and the other column would be dropped. AdultWeekend had the least null values and was kept. After all this, the data was reviewed for missing values which would be dealt with later as those rows may still have important information.

To recap, we now have a dataset that describes 277 resorts across 35 states with 24 features that can be used to predict price. We also have a seperate table that captures state level statistics and may be useful in predicting price. There are still 5 features with a significant amount of missing values, 4 of them are missing between 13 - 15% and 'NightSkiing_ac' is missing 41% of values.
